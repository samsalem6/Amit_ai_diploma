#### ""Problem: Basic Calculator""

##### Objective: Create a simple calculator program that can perform basic arithmetic operations: addition, subtraction, multiplication, and division.



""Requirements:""
------------------------
""User Input:""

The program should prompt the user to choose an operation from the following list:
- Addition
- Subtraction
- Multiplication
- Division
- Exit
Input Validation:

Ensure that the user inputs valid numbers for the calculations. If the input is invalid (e.g., non-numeric), prompt the user to enter the numbers again.
Performing Calculations:

Based on the user's choice, the program should ask for two numbers and perform the selected operation.
If the user selects division, check if the second number is zero to avoid division by zero errors.
Looping:

After each calculation, ask the user if they want to perform another calculation. If they answer "yes," return to the operation selection; if they answer "no," exit the program.
Output:

Display the result of the calculation in a user-friendly format.
For example: "The result of adding 5 and 3 is 8."
Exit Option:

If the user chooses to exit, print a thank you message and terminate the program.

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y
print("==================")

choices = {
    '1': (add, '+'),
    '2': (subtract, '-'),
    '3': (multiply, '*'),
    '4': (divide, '/')
}

choice = input("Enter choice for (1: add/2: subtract/3: multiply/4: divide): ")


# print(validate_choice(choice))
def calculation(choice, num1, num2):
    func, opr = choices[choice]
    result = func(num1, num2)
    print("============")
    return print(f"The result of {num1} {opr} {num2} = {result}")

def call_operations(choice):
    # validate if choice is in choices
    try:
        if choice in choices:
            num1 = input("Enter first number: ")
            num2 = input("Enter second number: ")

            # validate if num1 and num2 are numbers
            if not num1.isdigit() or not num2.isdigit():
                print("Invalid input, please enter numbers only")
                return call_operations(choice)
            else:
                # convert num1 and num2 to integers
                # and call the calculation function
                if choices[choice][0] == divide and int(num2) == 0:
                    print("Error: Division by zero is not allowed")
                else:
                    return calculation(choice, int(num1), int(num2))
                
        else:
            print("Invalid choice")
    except Exception as m:
        print(m)

# repeat the operation until user enters 'q'


call_operations(choice)
continue_choice = input("Do you want to continue? (yes/no) or (q) to exit: ")

def continue_calculation(continue_choice):
    # validate if continue_choice is in choices
    while True:
        if continue_choice.lower() == 'yes' or continue_choice.lower() == 'y':
            choice = input("Enter choice for (1: add/2: subtract/3: multiply/4: divide): ")
            call_operations(choice)
            ask_again = input("Do you want to continue? (yes/no) or (q) to exit: ")
            continue_calculation(ask_again)
            break
        elif continue_choice.lower() == 'no' or continue_choice.lower() == 'n' or continue_choice == 'q':
            if continue_choice == 'q':
                print("Exiting the calculator...")
            else:
                print("Thank you for using the calculator!")
            break
        else:
            print("Invalid choice")
            continue_choice = input("Do you want to continue? (yes/no): ")
            continue_calculation(continue_choice)
        
        
continue_calculation(continue_choice)
```

```python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        print("Error: Division by zero is not allowed.")
        return None
    return x / y

operations = {
    '1': (add, '+'),
    '2': (subtract, '-'),
    '3': (multiply, '*'),
    '4': (divide, '/')
}

def get_number(prompt):
    while True:
        value = input(prompt)
        try:
            return float(value)
        except ValueError:
            print("Invalid input. Please enter a valid number.")

def get_choice():
    print("\nSelect operation:")
    for key, (_, symbol) in operations.items():
        print(f"{key}: {symbol}")
    return input("Enter choice (1/2/3/4): ")

def perform_operation():
    choice = get_choice()
    if choice not in operations:
        print("Invalid choice.")
        return

    num1 = get_number("Enter first number: ")
    num2 = get_number("Enter second number: ")

    func, symbol = operations[choice]
    result = func(num1, num2)
    if result is not None:
        print(f"The result of {num1} {symbol} {num2} = {result}")

def run_calculator():
    while True:
        perform_operation()
        cont = input("Do you want to continue? (yes/no or q to quit): ").lower()
        if cont in ['no', 'n', 'q']:
            print("Thank you for using the calculator!")
            break

# Start the calculator
run_calculator()
```
